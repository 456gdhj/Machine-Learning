```python
import pandas as pd
```

```python
data={'catagory':['A','B','C','A','B','C']}
```

```python
df=pd.DataFrame(data)
df.head()
```

```python
one_hot_encoding=pd.get_dummies(df,columns=['catagory'])
one_hot_encoding.head()
```

```python
one_hot_encoding=pd.get_dummies(df,columns=['catagory'],prefix='Dummy')
one_hot_encoding.head()
```

```python
one_hot_encoding=pd.get_dummies(df,columns=['catagory'],prefix='Dummy',drop_first=True)
one_hot_encoding.head()
```
